## **Explanatory Analysis 2: Distribution of Sales Methods Used by Customers and Revenue Contribution**

In this section, we explore how customers engage with various sales methods, visualizing both the distribution of methods used and their corresponding revenue contributions. This analysis not only highlights customer preferences but also reveals the effectiveness of each approach in driving revenue, providing us with valuable insights to refine our sales strategies.

```python
# Set the figure size and create two subplots side by side
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Get the unique sales methods to maintain consistent colors
sales_methods = product_sales['sales_method'].unique()

# Use Seaborn's "Set2" palette to assign colors to each sales method
color_palette = sns.color_palette("Set2", n_colors=len(sales_methods))
color_dict = dict(zip(sales_methods, color_palette))

# --- First Plot: Revenue Distribution by Sales Method (Percentage) ---
# Calculate total revenue by sales method
revenue_by_method = product_sales.groupby('sales_method')['revenue'].sum().reset_index()

# Calculate the percentage of total revenue for each sales method
total_revenue = revenue_by_method['revenue'].sum()
revenue_by_method['percentage'] = (revenue_by_method['revenue'] / total_revenue) * 100

# Sort the DataFrame by percentage in descending order
revenue_by_method = revenue_by_method.sort_values(by='percentage', ascending=False)

# Plotting the bar chart with percentages on the first axis
axes[0].bar(revenue_by_method['sales_method'], revenue_by_method['percentage'], color=[color_dict[m] for m in revenue_by_method['sales_method']])

# Add a title and labels for the first plot with extra space below x-axis labels
axes[0].set_title('Revenue Distribution by Sales Method (Percentage)', fontsize=16)
axes[0].set_xlabel('Sales Method', fontsize=12, labelpad=10)  # Extra space with labelpad
axes[0].set_ylabel('Percentage of Total Revenue (%)', fontsize=12)

# Show the percentage value on top of each bar in the first plot
for i, value in enumerate(revenue_by_method['percentage']):
    axes[0].text(i, value + 1, f'{value:.1f}%', ha='center', fontsize=10)

# --- Second Plot: Customer Count by Sales Method ---
# Use countplot to visualize the number of customers for each sales method on the second axis
sns.countplot(x='sales_method', data=product_sales, palette=color_dict, order=product_sales['sales_method'].value_counts().index, ax=axes[1])

# Add text labels on the bars for the second plot
for p in axes[1].patches:
    axes[1].annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()), 
                     ha='center', va='baseline', fontsize=12, color='black')

# Customize the second plot with axes and black borders
axes[1].spines['top'].set_visible(True)
axes[1].spines['right'].set_visible(True)
axes[1].spines['left'].set_visible(True)
axes[1].spines['bottom'].set_visible(True)
for spine in axes[1].spines.values():
    spine.set_edgecolor('black')

# Draw x and y axes with ticks for the second plot
axes[1].axhline(0, color="black", linewidth=1)
axes[1].axvline(-0.5, color="black", linewidth=1)
axes[1].tick_params(axis='both', which='both', length=5, color='black')

# Set ticks for x and y axes on the second plot with no rotation on x-axis labels and extra label space
axes[1].set_xticks(range(len(product_sales['sales_method'].value_counts().index)))
axes[1].set_xticklabels(product_sales['sales_method'].value_counts().index)
axes[1].set_xlabel('Sales Method', fontsize=12, labelpad=10)  # Extra space with labelpad
axes[1].set_ylabel('Number of Customers', fontsize=12)
axes[1].yaxis.set_major_locator(plt.MaxNLocator(integer=True))

# Set titles and labels for the second plot
axes[1].set_title('Customer Count by Sales Approach', fontsize=14)

# Adjust layout for better spacing and presentation
plt.tight_layout()

# Show the plots side by side
plt.show()
```

**Key Insights:**

- **Email** emerges as the dominant sales method, with **7,465 customers** preferring this approach. This reflects its efficiency and convenience for both the customer and the sales team. Not only is it the most popular, but it also leads in revenue generation, contributing a substantial **51.3%** of total revenue. Clearly, email remains a powerful tool in our sales strategy.

- **Call** ranks second, with **4,961 customers** opting for more personalized, real-time interaction. While not as prevalent as email, this method plays an essential role, contributing **17.4%** to overall revenue. This suggests that direct contact still holds value in building customer relationships.

- The combination of **Email + Call**, used by **2,572 customers**, generates **31.3%** of total revenue. This blended approach shows the potential of integrating multiple communication strategies to maximize engagement and revenue.

---

**Strategic Implications:**

The data reveals that while **Email** is the most widely used and revenue-generating method, **Email + Call** stands out as a potent combination for driving results. Although calls alone are less effective, when integrated with email campaigns, they contribute significantly to overall sales.

This insight highlights the importance of refining our approach to customer outreach. Focusing on optimizing email campaigns while strategically incorporating calls for more personalized engagement can enhance revenue performance and strengthen customer relationships. By understanding these patterns, we can better allocate resources and tailor our sales strategies to meet customer preferences and maximize impact.

## **Explanatory Analysis 4: What Do the Revenue Trends Reveal About Our Sales Methods?**

To understand how each sales method contributes to revenue over time, we analyze the revenue trends across a six-week period for three distinct sales methods: Call, Email, and Email + Call. This line chart will reveal the fluctuations and patterns in revenue generation associated with each method, helping us identify which approaches are sustaining momentum, which may require optimization, and how each method performs relative to others.

By visualizing these trends, we gain insights into the stability, peaks, and shifts in revenue, enabling a data-driven evaluation of each method's effectiveness over time.

```python
# Set the Seaborn style for a cleaner look
sns.set_theme(style="white")  # Use 'white' for no grid lines

# Grouping data by week and sales method to calculate total revenue per method per week
revenue_over_time = product_sales.groupby(['week', 'sales_method'])['revenue'].sum().reset_index()

# Creating the line plot for weekly revenue trends
plt.figure(figsize=(14, 8))

# Using a more vibrant and professional color palette and bold lines
sns.lineplot(data=revenue_over_time, x='week', y='revenue', hue='sales_method', 
             marker='o', palette='deep', linewidth=3, style='sales_method')

# Enhancing the title and labels with larger font sizes
plt.title('Weekly Revenue Trends by Sales Method', fontsize=24, fontweight='bold')
plt.xlabel('Week', fontsize=18, fontweight='bold')
plt.ylabel('Total Revenue', fontsize=18, fontweight='bold')

# Customizing the y-axis ticks for better clarity
plt.yticks(fontsize=14)  # Increase font size for y-axis ticks
plt.xticks(range(product_sales['week'].min(), product_sales['week'].max() + 1), fontsize=14)

# Adding black borders to the figure
plt.gca().spines['top'].set_color('black')
plt.gca().spines['top'].set_linewidth(1.5)
plt.gca().spines['right'].set_color('black')
plt.gca().spines['right'].set_linewidth(1.5)
plt.gca().spines['left'].set_color('black')
plt.gca().spines['left'].set_linewidth(1.5)
plt.gca().spines['bottom'].set_color('black')
plt.gca().spines['bottom'].set_linewidth(1.5)

# Positioning the legend on the right and adjusting its style
plt.legend(title='Sales Method', title_fontsize='15', fontsize='14', loc='center left', bbox_to_anchor=(1, 0.5))
plt.tight_layout()

# Display the plot
plt.show()
```

### **Insight:**

The chart provides a clear illustration of the revenue trends over a six-week period for the different sales methods: **Call**, **Email**, and **Email + Call**. 

1. **Overall Revenue Performance:**
   - The general trend shows a decrease in total revenue over time, which is a cause for concern. This downward trajectory suggests that the initial boost seen in Week 1 may not have been sustained, necessitating a review of sales strategies or market conditions.

2. **Sales Method Performance:**
   - **Email:** Initially, the Email method was the leading revenue generator, peaking in Week 1. However, it experienced a sharp decline in Weeks 2 and 3, with only minor stabilization in later weeks. This indicates that while Email is an effective initial contact method, it may require additional support or follow-up to maintain momentum.
   - **Call:** The Call method consistently underperformed compared to the other methods. Despite slight increases, its revenue contribution remained the lowest throughout the period. This consistent underperformance suggests that calls alone may not be an efficient or impactful sales approach.
   - **Email + Call:** The combined Email + Call method demonstrated steady growth, ultimately surpassing the other methods in revenue generation by Week 6. This consistent upward trend highlights the strength of a multifaceted approach, suggesting that combining methods can create a synergistic effect, leading to more robust sales performance.

3. **Strategic Recommendation:**
   - Given the data, it is clear that the **Email + Call** method is the most effective strategy, particularly in sustaining and growing revenue over time. This approach should be prioritized and potentially expanded. 
   - In contrast, the **Email** method's decline indicates a need for strategy refinement, possibly through better targeting, content optimization, or integration with other methods.
   - **Call** strategies, given their underperformance, might be re-evaluated or used more selectively, potentially in combination with email campaigns to boost effectiveness.

   - **Monitor Trends:** Ongoing monitoring of these trends is essential. Continuous assessment will allow for timely adjustments and ensure that the sales strategies remain aligned with market responses.
